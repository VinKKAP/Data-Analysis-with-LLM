```python
def rename_columns(df):
    """
    Rename columns to more descriptive, standardized names.
    
    Args:
        df (pandas.DataFrame): Input dataframe to rename columns
    
    Returns:
        pandas.DataFrame: Dataframe with renamed columns
    """
    # Create a comprehensive mapping dictionary
    column_mapping = {
                        # Demographic
                        'sex': 'sex',
                        'age': 'age',
                        'birthyr': 'year_of_birth',
                        'race': 'race',
                        'raced': 'detailed_race',
                        'hispan': 'hispanic_origin',
                        'hispand': 'detailed_hispanic_origin',
                        'bpl': 'birthplace',
                        'bpld': 'detailed_birthplace',
                        'ancestr1': 'first_ancestry',
                        'ancestr1d': 'detailed_first_ancestry',
                        'citizen': 'citizenship_status',
                        'yrnatur': 'year_of_naturalization',
                        'yrimmig': 'year_of_immigration',
                        'yrsusa1': 'years_in_usa',
                        'speakeng': 'english_proficiency',
                        'racamind': 'american_indian_race',
                        'racasian': 'asian_race',
                        'racblk': 'black_race',
                        'racpacis': 'pacific_islander_race',
                        'racwht': 'white_race',
                        'racother': 'other_race',

                        # Household
                        'nchlt5': 'child_age_leq_4',
                        'famsize': 'family_size',
                        'nchild': 'number_of_children',
                        'eldch': 'age_of_eldest_child',
                        'yngch': 'age_of_youngest_child',
                        'nsibs': 'number_of_siblings',
                        'prent': 'parent_presence',
                        'hhwt': 'household_weight',
                        'metro': 'metropolitan_status',
                        'region': 'us_census_region',
                        'gq': 'group_quarters_status',

                        # Marital and Fertility
                        'marst': 'marital_status',
                        'marrno': 'number_of_marriages',
                        'marrinyr': 'married_within_year',
                        'yrmarr': 'year_of_marriage',
                        'fertyr': 'fertility_year',

                        # Education
                        'school': 'school_attendance',
                        'educ': 'education',
                        'educd': 'detailed_education',
                        'gradeatt': 'grade_attended',
                        'gradeattd': 'detailed_grade_attended',
                        'schltype': 'school_type',
                        'degfield': 'degree_field',
                        'degfieldd': 'detailed_degree_field',
                        'degfield2': 'second_degree_field',
                        'degfield2d': 'detailed_second_degree_field',
                        'edscor90': 'education_score_1990',

                        # Employment
                        'empstat': 'employment_status',
                        'empstatd': 'detailed_employment_status',
                        'labforce': 'labour_force_participation',
                        'occ': 'occupation',
                        'occ1990': 'occupation_1990_classification',
                        'occ2010': 'occupation_2010_classification',
                        'ind': 'industry',
                        'ind1990': 'industry_1990_classification',
                        'classwkr': 'class_of_worker',
                        'classwkrd': 'detailed_class_of_worker',
                        'occsoc': 'soc_occupation',
                        'indnaics': 'naics_industry',
                        'wkswork2': 'weeks_worked',
                        'uhrswork': 'usual_hours_worked',
                        'workedyr': 'worked_last_year',
                        'incwage': 'wage_income',

                        # Survey Metadata
                        'year': 'survey_year',
                        'datanum': 'dataset_number',
                        'serial': 'household_serial_number',
                        'pernum': 'person_number_within_household',
                        'perwt': 'person_weight',

                        # Veteran Status
                        'vetstat': 'veteran_status',
                        'vetstatd': 'detailed_veteran_status'
                    }
    
    # Rename the columns
    df_renamed = df.rename(columns=column_mapping)
    
    return df_renamed

# Call the function and print the result
df_renamed = rename_columns(df)
print(df_renamed)
```
